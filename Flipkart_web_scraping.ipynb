```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as uReq 
```

```python
my_url="https://www.flipkart.com/search?q=samsung+mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_4_na_na_na&as-pos=1&as-type=RECENT&suggestionId=samsung+mobiles%7CMobiles&requestId=d8e327db-02b3-4f15-ab81-d9c241ba83d7&as-backfill=on"
```

```python
uClient = uReq(my_url)
page_html = uClient.read()
uClient.close()
```

```python
page_html
```

```python
page_soup = soup(page_html, "html.parser")
```

```python
page_soup
```

```python
container = page_soup.findAll('div', {"class":"_1YokD2 _3Mn1Gg"})
```

```python
container
```

```python
print(soup.prettify(container[0]))
```

```python
containers = container[0]
containers
```

```python
product_name = containers.findAll('div',{'class':'_4rR01T'})
```

```python
product_name
```

```python
mobile_name = []
for i in product_name:
    mobile_name.append(i.text)
```

```python
mobile_name
```

```python
product_price = containers.findAll('div',{'class':'_30jeq3 _1_WHN1'})
```

```python
product_price
```

```python
mobile_price = []
for i in product_price:
    mobile_price.append(i.text)
```

```python
mobile_price
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list(zip(mobile_name,mobile_price)), columns=['mobile_name','mobile_price'])
```
